##Customer Segmentation

```python
%sql
SELECT
    CASE
        WHEN LifetimeRevenue >= 500000
        THEN 'PLATINUM'
        WHEN LifetimeRevenue >= 250000
        THEN 'GOLD'
        WHEN LifetimeRevenue >= 100000
        THEN 'SILVER'
        ELSE 'BRONZE'
    END AS Segment,

    COUNT(*) Customers,
    SUM(LifetimeRevenue) Revenue

FROM workspace.final.gold_customer_value

GROUP BY 1
```
